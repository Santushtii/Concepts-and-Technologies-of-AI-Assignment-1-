```python
print("\n6. ADDING HDI CATEGORY COLUMN")
print("")

"""Categorize HDI value into UNDP categories"""
def categorize_hdi(value):
        """Categorize HDI value into UNDP categories"""
        if pd.isna(value):
            return 'Unknown'
        elif value < 0.550:
            return 'Low'
        elif 0.550 <= value < 0.700:
            return 'Medium'
        elif 0.700 <= value < 0.800:
            return 'High'
        else:  # value >= 0.800
            return 'Very High'
        return df

# Apply categorization
df['HDI Category'] = df['hdi'].apply(categorize_hdi)

# Verify categorization
print("\nHDI Category Distribution:")
category_counts = df['HDI Category'].value_counts()
for category, count in category_counts.items():
    print(f"  {category}: {count} countries")

# Save final dataframe
output_path = 'HDI_category_added.csv'
df.to_csv(output_path, index=False)
print(f"\nFinal dataframe saved as: {output_path}")


```

```python
# Task 2: Composite Development Score
print("2. COMPOSITE DEVELOPMENT SCORE")
print("-")

# Check if required columns exist
required_cols = ['life_expectancy', 'gross_inc_percap']
missing_cols = [col for col in required_cols if col not in df.columns]

if missing_cols:
    print(f"Missing columns for composite score: {missing_cols}")

# Create Composite Score
df['Composite Score'] = (
    0.30 * df['life_expectancy'] +
    0.30 * df['gross_inc_percap']
)

# Rank countries by Composite Score
df = df.sort_values('Composite Score', ascending=False)
df['Composite Rank'] = df['Composite Score'].rank(ascending=False, method='min').astype(int)

df

print("\nSouth Asian Countries Ranked by Composite Score(2022):")
for i, (idx, row) in enumerate(df.iterrows(), 1):
    print(f"{i:2d}. {row['country']:15s} Composite: {row['Composite Score']:.3f} "
          f"HDI: {row.get('hdi', 'N/A'):.3f}")

# Plot top 5 countries
top_5 = df.head(5)

plt.figure(figsize=(10, 6))
bars = plt.barh(top_5['country'], top_5['Composite Score'],
                color='teal', edgecolor='black')

plt.title('Top 5 South Asian Countries by Composite Score', fontsize=14, fontweight='bold')
plt.xlabel('Composite Score', fontsize=12)
plt.xlim(0, max(top_5['Composite Score']) * 1.1)

# Add value labels
for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height()/2,
            f'{width:.3f}', ha='left', va='center', fontweight='bold')

plt.gca().invert_yaxis()  # Highest at top
plt.tight_layout()
plt.savefig('visualizations/composite_score_top5.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nComparison with HDI ranking (if available):")
if 'hdi' in df.columns:
    df['HDI Rank'] = df['hdi'].rank(ascending=False, method='min')

    comparison = df[['country', 'Composite Score', 'Composite Rank',
                                    'hdi', 'HDI Rank']].copy()
    comparison['Rank Difference'] = comparison['HDI Rank'] - comparison['Composite Rank']

    print("\nCountries where Composite Score ranking differs from HDI ranking:")
    diff_countries = comparison[comparison['Rank Difference'] != 0]
    if len(diff_countries) > 0:
        for _, row in diff_countries.iterrows():
            direction = "higher" if row['Rank Difference'] > 0 else "lower"
            print(f"  {row['country']}: Composite rank is {abs(row['Rank Difference'])} positions {direction}")
    else:
        print("  No differences in ranking")
```

```python
# Outlier detection
print("3. OUTLIER DETECTION")
print("-")

import matplotlib.pyplot as plt

def detect_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    return (series < Q1 - 1.5 * IQR) | (series > Q3 + 1.5 * IQR)
```

```python
df["HDI_outlier"] = detect_outliers(df["hdi"])
df["GNI_outlier"] = detect_outliers(df["gross_inc_percap"])
```

```python
plt.figure()
plt.scatter(df["gross_inc_percap"], df["hdi"])
plt.scatter(
    df[df["HDI_outlier"] | df["GNI_outlier"]]["gross_inc_percap"],
    df[df["HDI_outlier"] | df["GNI_outlier"]]["hdi"]
)
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("HDI vs GNI per Capita (Outliers Highlighted)")
plt.show()

print("Scatter plot with outliers highlighted")
print("\n No outliers detected")
print("\n Indicates consistent development and no extremities in the south asian countries")
```

```python
# HDI Disparity

print("HDI Disparity: ")

def hdi_stats(df):
    '''
    Computes Range and Coefficient of Variance and indentifies HDI disparity
    '''
    mean = df["hdi"].mean()
    std = df["hdi"].std()
    hdi_range = df["hdi"].max() - df["hdi"].min()
    cv = std / mean
    return mean, std, hdi_range, cv

# Compute statistics
sa_mean, sa_std, sa_range, sa_cv = hdi_stats(df_sa_2020)
me_mean, me_std, me_range, me_cv = hdi_stats(df_me_2020)

print(f"South Asia: \nRange: {sa_range:.4f}, CV: {sa_cv:.4f}")
print(f"\nMiddle East: \nRange: {me_range:.4f}, CV: {me_cv:.4f}")

# Identify region with more variation
if sa_cv > me_cv:
    print("\nSouth Asia exhibits more HDI variation.")
else:
    print("\nMiddle East exhibits more HDI variation.")

```

```python
# Outlier Detection
print("Outlier Detection ")

def detect_outliers_iqr(series):
    '''
    Another method to find out outliers using 1.5 Ã— IQR rule.
    '''
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return (series < lower) | (series > upper)

# Create regional datasets using .copy()
df_sa_2020 = df_sa_2020.copy()
df_me_2020 = df_me_2020.copy()

# Outliers for South Asia:
df_sa_2020["HDI_outlier"] = detect_outliers_iqr(df_sa_2020["hdi"])
df_sa_2020["GNI_outlier"] = detect_outliers_iqr(df_sa_2020["gross_inc_percap"])

df_sa_2020["outlier"] = df_sa_2020["HDI_outlier"] | df_sa_2020["GNI_outlier"]

#Outliers for Middle East:
df_me_2020["HDI_outlier"] = detect_outliers_iqr(df_me_2020["hdi"])
df_me_2020["GNI_outlier"] = detect_outliers_iqr(df_me_2020["gross_inc_percap"])

df_me_2020["outlier"] = df_me_2020["HDI_outlier"] | df_me_2020["GNI_outlier"]

# List outlier values explicitly
df_me_2020.loc[df_me_2020["outlier"],
               ["country", "year", "hdi", "gross_inc_percap"]]


```
